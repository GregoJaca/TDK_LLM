```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
n_points = 150   # number of perturbed initial conditions
radius = 0.15    # radius of perturbation sphere (smaller)
vector_length = 2.0  # length of the original vector

# Define the original initial condition vector
original = np.array([0, 0.5, vector_length])

# Generate perturbed points: uniformly random on a sphere
directions = np.random.normal(size=(n_points, 3))
directions /= np.linalg.norm(directions, axis=1)[:, None]  # normalize to unit vectors
perturbations = directions * radius
points = original + perturbations

# Plot
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# Perturbed points
ax.scatter(points[:, 0], points[:, 1], points[:, 2],
           s=10, color="royalblue", alpha=0.4, label="Perturbed initial conditions")
ax.text(original[0], original[1] + 0.2, original[2] + 0.16,
        "Perturbed initial conditions", fontsize=10, color="royalblue")

# Original vector as an arrow
ax.quiver(0, 0, 0, original[0], original[1], original[2],
          color="crimson", linewidth=2.5, arrow_length_ratio=0.1)
ax.text(original[0], original[1], original[2] - 0.4,
        "Original initial condition", fontsize=10, color="crimson")



# Tripod-style axes from origin
axis_len = 2.0
ax.quiver(0, 0, 0, axis_len, 0, 0, color="black", arrow_length_ratio=0.05)
ax.quiver(0, 0, 0, 0, axis_len, 0, color="black", arrow_length_ratio=0.05)
ax.quiver(0, 0, 0, 0, 0, axis_len, color="black", arrow_length_ratio=0.05)
ax.text(axis_len, 0.1, 0, "X", fontsize=10)
ax.text(0, axis_len, 0, "Y", fontsize=10)
ax.text(0, 0, axis_len, "Z", fontsize=10)

# View and style
ax.set_xlim([0, axis_len])
ax.set_ylim([0, axis_len])
ax.set_zlim([0, axis_len])
ax.view_init(elev=25, azim=35)
ax.set_axis_off()

plt.tight_layout()
plt.show()
```
